# Explore HAR with Smartphone Dataset

https://www.kaggle.com/datasets/uciml/human-activity-recognition-with-smartphones/data 

The dataset captures 3-axial linear acceleration and 3-axial angular velocity from the accelerometer and gyroscope of a waist-mounted smartphone. It aims to recognize human activities, classifying them into six categories.
Features:

- Triaxial acceleration (total and estimated body acceleration)
- Triaxial angular velocity
- A 561-feature vector with time and frequency domain variables
    
  


```python
TRAIN_FILE = '../../datafiles/har_with_smartphones/train.csv'
TRAIN_FILE = Path(TRAIN_FILE)
```

```python
train_df = pd.read_csv(TRAIN_FILE)
train_df.head(10)
```

<hr>
Labels:

```python
print(len(list(train_df.Activity.unique())), list(train_df.Activity.unique()))
```

<hr>
For each subject and activity, you can create separate time series.

```python
pivot = train_df.set_index(['subject', 'Activity'])
pivot = pivot.groupby(['subject', 'Activity']).size().unstack(fill_value=0)
pivot['Total'] = pivot.sum(axis=1)
pivot.loc['Total'] = pivot.sum()
pivot
```
